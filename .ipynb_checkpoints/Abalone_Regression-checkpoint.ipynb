# Abalone Years Predictive Model

**Notes:** 
* Use 42 as your seed for 15% for unseen
* Use 42 as your seed for setup()
* Predict Years
* Use a sklearn regressor- double check by running models()
* Use sklearn.metrics and find the r2,mae,mse, rmse

```python
import numpy as np
import pandas as pd
import udf
```

```python
data = pd.read_csv('./data/abalone.csv')
data
```

```python
df =data.drop(['Rings'], axis=1)
```

```python
data, unseen = udf.gen_unseen_pct(data=df, frac=0.15, random_state=42)
df.shape, data.shape, unseen.shape
```

```python
data.head()
```

```python
# import RegressionExperiment and init the class
from pycaret.regression import RegressionExperiment
s = RegressionExperiment()
```

```python
s.setup(data, target = 'Years', session_id = 42, ignore_features=['Rings'], categorical_features=['Sex'])
```

```python
# check the type of exp
type(s)
```

```python
s.compare_models()
```

```python
gbr = s.create_model('gbr')
```

```python
tuned_gbr = s.tune_model(gbr)
```

```python
final_gbr = s.finalize_model(tuned_gbr)
```

```python
final_gbr
```

```python
s.evaluate_model(gbr)
```

```python
df_pred = s.predict_model(final_gbr,unseen)
```

```python
df_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
```

```python
r2 = r2_score(df_pred['Years'], df_pred['prediction_label'])
mae = mean_absolute_error(df_pred['Years'], df_pred['prediction_label'])
mse = mean_squared_error(df_pred['Years'], df_pred['prediction_label'])
rmse = root_mean_squared_error(df_pred['Years'], df_pred['prediction_label'])
print (f'R2: {r2:2f}\nMAE: {mae:2f}\nMSE:{mse:2f}\nRMSE:{rmse:.2f}')
```
