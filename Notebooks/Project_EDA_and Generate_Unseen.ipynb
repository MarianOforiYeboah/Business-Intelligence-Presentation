```python
data = pd.read_csv('./data/smmh_main.csv',low_memory=False)
```

```python
from pycaret.classification import ClassificationExperiment
from imblearn.over_sampling import SMOTE

# Calculate the minimum number of samples
min_samples = min(data['Mental_Health_Disorders'].value_counts())

# Initialize SMOTE with ratio set to 'minority'
s = ClassificationExperiment()
s.setup(data, target='Mental_Health_Disorders', session_id=42, fix_imbalance=True, fix_imbalance_method='SMOTE')

```

```python
s.compare_models()
```

```python
lr = s.create_model('lr')
lr
```

```python
tuned_lr = s.tune_model(lr)   
```

```python
s.evaluate_model(tuned_lr) 
```

```python
final_lr = s.finalize_model(lr)
```

```python
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
unseen = pd.read_csv('./data/unseen_data.csv')
unseen
```

```python
cls = final_lr
unseen_X = unseen.drop('Mental_Health_Disorders', axis=1).reset_index(drop=True)
y = unseen['Mental_Health_Disorders']
```

```python
unseen_X.shape, y.shape
```

```python
y_pred = cls.predict(unseen_X)
y_pred
```

```python
print(classification_report(y, y_pred))
```

```python
accuracy_score(y, y_pred)
```

```python
precision_score(y, y_pred, pos_label='Positive', average='weighted')
```

```python
recall_score(y, y_pred, pos_label='Negative', average='weighted')
```
