```python
import numpy as np
import pandas as pd
import udf
```

```python
data = pd.read_csv('./data/SmMh_cleaned.csv')
data
```

```python
# First, we create a new column 'time_on_sm' in the DataFrame based on 'Daily_Time_Spent'
conditions = [
    (data['Daily_Time_Spent'] == 'More than 5 hours'),
    (data['Daily_Time_Spent'] == 'Between 4 and 5 hours'),
    (data['Daily_Time_Spent'] == 'Between 3 and 4 hours'),
    (data['Daily_Time_Spent'] == 'Between 2 and 3 hours'),
    (data['Daily_Time_Spent'] == 'Between 1 and 2 hours')
]
values = [6, 5, 4, 3, 2]

data['Time_Spent_on_SM'] = np.select(conditions, values, default=1)
data
```

```python
df =data.drop(['Id.1'], axis=1)
df
```

```python
df =data.drop(['Mental_Health_Disorders'], axis=1)
```

```python
data, unseen = udf.gen_unseen_pct(data=df, frac=0.2, random_state=42)
df.shape, data.shape, unseen.shape
```

```python
data.head()
```

```python
# import RegressionExperiment and init the class
from pycaret.regression import RegressionExperiment
s = RegressionExperiment()
```

```python
s.setup(data=df, 
        target='Time_Spent_on_SM', 
        session_id=42, 
        ignore_features=['Mental_Health_Disorders','Id', 'Id.1', 'Timestamp'], 
        categorical_features=['Gender', 'Relationship_Status', 'Occupation', 'Affiliated_Organizations', 'Social_Media_User', 'Platforms_Used'])
```

```python
# check the type of exp
type(s)
```

```python
s.compare_models()
```

```python
lgbm = s.create_model('lightgbm')
```

```python
print("Feature Names:", df.columns)
```

```python
# Replace special characters in feature values
for column in df.select_dtypes(include=['object']).columns:
    df[column] = df[column].str.replace('[^A-Za-z0-9]+', '_', regex=True)
```

```python
tuned_lgbm = s.tune_model(lgbm)
```

```python
final_lgbm = s.finalize_model(tuned_lgbm)
final_lgbm
```

```python
s.evaluate_model(lgbm)
```

```python
df_pred = s.predict_model(final_lgbm,unseen)
```

```python
df_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
```

```python
r2 = r2_score(df_pred['Years'], df_pred['prediction_label'])
mae = mean_absolute_error(df_pred['Years'], df_pred['prediction_label'])
mse = mean_squared_error(df_pred['Years'], df_pred['prediction_label'])
rmse = root_mean_squared_error(df_pred['Years'], df_pred['prediction_label'])
print (f'R2: {r2:2f}\nMAE: {mae:2f}\nMSE:{mse:2f}\nRMSE:{rmse:.2f}')
```
